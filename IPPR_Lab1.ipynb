```python
#read and display image
import cv2
image = cv2.imread('snd.jpg');
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print('Original Image');
```

```python
#convert a color image to grayscale omage
import cv2
image1 = cv2.imread('snd.jpg',0);
from google.colab.patches import cv2_imshow
cv2_imshow(image1)
print('Grayscale Image');
```

```python
#convert  an iamge to number of arrays
import cv2
image = cv2.imread('snd.jpg');
from google.colab.patches import cv2_imshow
cv2_imshow(image)
from numpy import asarray
data = asarray(image)
print(data);
```

```python
#image rotation
import cv2
import imutils
image = cv2.imread('snd.jpg');
from google.colab.patches import cv2_imshow
cv2_imshow(image)
R1 = imutils.rotate(image,90)
cv2_imshow(R1)
print('Original Image');
```

```python
#conversion of RGB model to CMY model
import cv2
image = cv2.imread('snd.jpg');
from google.colab.patches import cv2_imshow
R,G,B = cv2.split(image)
C = 255-R
M = 255 - G
Y = 255 - B
cv2_imshow(R)
print('Red Component')
cv2_imshow(C)
print('Cyan component')
```
